# MOST POPULAR DOGS

For the visualisation part of the project, I present the most popular dogs (by favorite/retweet count) by categories

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import Image
%matplotlib inline
from collections import Counter
import re
```

```python
master_df = pd.read_csv('twitter_archive_master.csv')
```

```python
max_retweet_id = master_df['retweet_count'].idxmax()
max_favorite_id = master_df['favorite_count'].idxmax()
print('Max retweets ID: ' + str(max_retweet_id))
print('Max favourites ID: ' + str(max_retweet_id))
```

## MOST POPULAR DOG - OVERALL

```python
print('The most popular dog overall: ' + str(master_df.name[max_retweet_id]))
print('Appeared on we rate dogs on: ' + str(master_df.timestamp[max_retweet_id]))
print('Classified as: ' + str(master_df.level[max_retweet_id]))
print('Retweet Count: ' + str(master_df.retweet_count[max_retweet_id]))
print('Favourite Count: ' + str(master_df.favorite_count[max_retweet_id]))
print('Rated: ' + str(master_df.rating_numerator[max_retweet_id]) + '/' + str(master_df.rating_denominator[max_retweet_id]))
print('NN best prediction: ' + str(master_df.p1[max_retweet_id]) + ' with ' + str(round(master_df.p1_conf[max_retweet_id], 3)*100) + '% confidence')
print('Tweet: ' + re.sub(r"http\S+", "", master_df.full_text[max_retweet_id]))
Image(master_df.jpg_url[max_retweet_id])
```
